```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices.csv')
print("Dataset:\n", df)
```

```python
plt.figure(figsize=(8, 5))
plt.xlabel('Area (sq ft)', fontsize=12)
plt.ylabel('Price (PKR)', fontsize=12)
plt.scatter(df.area, df.price, color='red', marker='+', label='Data Points')
plt.legend()
plt.title('Area vs Price', fontsize=14)
plt.show()
```

```python
X = df[['area']] 
y = df['price']   
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
m = model.coef_[0] 
b = model.intercept_  
print(f"Equation of line: Price = {m} * Area + {b}")
```

```python
predicted_prices = model.predict(X)
df['predicted_price'] = predicted_prices
print("\nTask 1 - Dataset with Predicted Prices:\n", df)
```

```python
plt.figure(figsize=(8, 5))
plt.xlabel('Area (sq ft)', fontsize=12)
plt.ylabel('Price (PKR)', fontsize=12)
plt.scatter(df.area, df.price, color='red', marker='+', label='Actual Prices')
plt.plot(df.area, predicted_prices, color='blue', label='Regression Line')
plt.legend()
plt.title('Regression Line for Dataset', fontsize=14)
plt.show()
```
