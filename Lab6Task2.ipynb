```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
X = df[['area']] 
y = df['price'] 
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
new_areas = pd.DataFrame([[5000], [8000], [9000]], columns=['area'])  
new_predictions = model.predict(new_areas)
```

```python
print("Task 2 - Predicted Prices for New Areas:")
for area, price in zip(new_areas['area'], new_predictions):
    print(f"Area: {area} sq ft -> Predicted Price: {price:.2f} PKR")
```

```python
plt.figure(figsize=(8, 5))
plt.xlabel('Area (sq ft)', fontsize=12)
plt.ylabel('Price (PKR)', fontsize=12)
plt.scatter(df.area, df.price, color='red', marker='+', label='Actual Prices') 
plt.plot(X, model.predict(X), color='blue', label='Regression Line')  
plt.scatter(new_areas['area'], new_predictions, color='green', marker='x', s=100, label='New Predictions')
plt.legend()
plt.title('Prediction for New Areas', fontsize=14)
plt.show()
```
